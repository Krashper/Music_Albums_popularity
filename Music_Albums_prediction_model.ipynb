We shouldn't delete any rows with missings, cause they are single songs and don't have other songs, so there are missing only columns about song2 and song3 in these rows.

I think we can combine the features of 3 songs into one.

## Delete rows with missing features of song_0

```python
data = data[data["t_dance0"].notna()]

data.isna().sum()
```

## Сonverting 3 features into 1

```python
cols = ["t_dur", "t_dance", "t_energy", "t_key", "t_mode", "t_speech", "t_acous", "t_ins", "t_live", "t_val", "t_tempo", "t_sig"]

new_cols = []

for col in cols:
    new_col_name = f"{col}_avg"
    data[new_col_name] = data[[f"{col}0", f"{col}1", f"{col}2"]].mean(axis=1)
    
    new_cols.append(new_col_name)
```

```python
data
```

```python
data[["t_mode0", "t_mode1", "t_mode2"]][data["id"] == 159996] # checking the correctness
```

# EDA of the converted dataset

```python
cols = new_cols + ["release_date", "total_tracks", "popularity"]

data_conv = data[cols]
```

```python
data_conv.columns
```

## Scatterplots

```python
X = data_conv.drop(columns=["popularity"])

Y = data_conv["popularity"]
```

```python
for X_col in X.columns:
    plt.scatter(X[X_col], Y, c="blue")
    plt.title(f"Scatter plot with {X_col}")
    plt.show()
```
